```python
%matplotlib inline
import matplotlib as mlt
import matplotlib.pyplot as plt
import mpld3
from mpld3 import plugins
import numpy as np
mlt.style.use('ggplot')
mpld3.enable_notebook()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
X = range(-10,11)

plt.plot(X, [a*2 for a in X], marker='o', label='y=2x')
plt.plot(X, [a**2 for a in X], marker='+', label='y=x^2')
plt.legend(loc='upper right')
plt.title('plot title')
plt.xlabel('x')
plt.ylabel('y')
```

```python
X = np.linspace(-10, 10, 30)
plt.plot(X, [1.0/a for a in X], marker='*')
```

```python
import sympy as sym
x = sym.Symbol('x')
sym.plot(1.0/x+3)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(6, 6))
fig.subplots_adjust(hspace=0.1, wspace=0.1)
ax = ax[::-1]

X = np.random.normal(size=(3, 100))
for i in range(3):
    for j in range(3):
        ax[i, j].xaxis.set_major_formatter(plt.NullFormatter())
        ax[i, j].yaxis.set_major_formatter(plt.NullFormatter())
        points = ax[i, j].scatter(X[j], X[i])

plugins.connect(fig, plugins.LinkedBrush(points))
```
