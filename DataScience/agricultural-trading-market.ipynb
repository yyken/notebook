```python
import os
import re
import pandas as pd
```

## Read all json file

```python
json_files = [ f for f in os.listdir('data/ATM-vegetable-2013.json/') ]
fn_re = re.compile(r"^(\d{3}-\d{2}-\d{2})-(\d{3})")
```

```python
def read_data(fn):
    g = fn_re.findall(fn);
    dt = pd.read_json('data/ATM-vegetable-2013.json/' + fn)
    dt['date'] = g[0][0]
    dt['area'] = g[0][1]
    return dt
```

先讀一個檔案試試

沒問題後，一次把所有的檔案讀進來，時間會有點久

```python
df = pd.DataFrame()
for fn in json_files:
    d = read_data(fn)
    df = df.append(d)
```
